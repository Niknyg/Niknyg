# Plotting and visualization

We are going to use the [Matplotlib]() and [Seaborn]() library to produce visualizations for our data.

**Matplotlib** is probably the best-known and most used Python visualization library, and **Seaborn** is a fancy and easy-to-use wrapper around Matplotlib's functions that (i) provides extra plot types, and (ii) usually reduces the number of lines of code you need to use to obtain the same plot.

However, matplotlib allows you to more granularly set the aesthetics of your plot.

Therefore, it is extremely common to use them as a combo ðŸ˜„

```python
# Import modules from 'matplotlib'.
import matplotlib.pyplot as plt

# Import modules from 'seaborn'.
import seaborn as sns
```

```python
# Example dataset: health expenditure & life expectancy for a few
# countries.
#
# Columns: Year, Country, Spending_USD, Life_Expectancy
healthexp = sns.load_dataset("healthexp").sort_values("Year")

# Inspect the first lines of the data frame.
healthexp.head()
```

```python
# Second example dataset: tips.
tips = sns.load_dataset("tips")

# Inspect the first lines of the data frame.
tips.head()
```

## Line plots

Line plots connect data points with straight line segments, showing how a variable changes over a continuous axis (often time). They are ideal for **trends**, **evolution over time**, and **comparing multiple series** on the same scale (e.g. life expectancy for several countries). Line plots work best when the x-axis is ordered and when you care about how values move between points, not just their individual magnitudes.

### Basic line and color

References:

- The [`seaborn.lineplot`](https://seaborn.pydata.org/generated/seaborn.lineplot.html) function.
- The [`matplotlib.pyplot.subplots`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html) function.
- The [`matplotlib.axes.Axes.set_title`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html) function.
- The [`matplotlib.axes.Axes.set_ylabel`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylabel.html) function.

```python
# Create a list with only some countries of interest.
countries = ["USA", "Japan", "Germany"]

# Select a portion of our data frame with only data regarding the
# chosen countries.
df = healthexp[healthexp["Country"].isin(countries)]

# Use the 'matplotlib.pyplot.subplots' function to create:
# - The 'figure' (fig), aka the full plot area.
# - The 'axis' (ax), which, a bit counterintuitively, refers to a
#   single panel within the figure (especially important for when you
#   have multiple panels) and NOT to the x- or y-axis.
fig, ax = plt.subplots()

# Create a simple line plot using the 'seaborn.lineplot' function.
sns.lineplot(

    # Set the input data frame.
    data = df,

    # Set the name of the column containing the data we want displayed
    # on the x-axis.
    x = "Year",

    # Set the name of the column containing the data we want displayed
    # on the y-axis.
    y = "Life_Expectancy",

    # Set the name of the column containing the data we want to use to
    # group the data by color.
    hue = "Country",

    # Set on which 'axis' (panel) we want to draw the plot.
    ax = ax,
)

# Set the title of the plot.
ax.set_title("Life expectancy over time")

# Set the label of the y-axis.
ax.set_ylabel("Life expectancy (years)")

# Call the 'matplotlib.pyplot.show' function to display our plot.
plt.show()

# You can see that the name of the x-axis is the name of the column
# containing the data displayed on the axis.

# You can also see that a legend is automatically displayed.
```

### Custom color palette

We can also use a custom color palette, choosing which color each line is going to be displayed in.

The palette can be, for instance:

- A dictionary mapping the values of the variable we want to use for coloring to the colors that should be use for each or them.
- A string representing the name of a color palette among those available in either Matplotlib or Seaborn.

References:

- The [`seaborn.lineplot`](https://seaborn.pydata.org/generated/seaborn.lineplot.html) function.
- The [`matplotlib.pyplot.subplots`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html) function.
- The [`matplotlib.axes.Axes.set_title`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html) function.
- The [`matplotlib.axes.Axes.set_ylabel`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylabel.html) function.
- A list of [all markers available in matplotlib](https://matplotlib.org/stable/api/markers_api.html).
- The color palettes available in [Matplotlib](https://matplotlib.org/stable/users/explain/colors/colormaps.html) or [Seaborn](https://seaborn.pydata.org/generated/seaborn.color_palette.html#seaborn.color_palette).

```python
# Define the custom palette as a dictionary mapping each country to
# the color their corresponding line will have.
custom_palette = {
    "USA": "tab:pink",
    "Japan": "tab:orange",
    "Germany": "tab:green",
}

# Use the 'matplotlib.pyplot.subplots' function to create:
# - The 'figure' (fig), aka the full plot area.
# - The 'axis' (ax), which, a bit counterintuitively, refers to a
#   single panel within the figure (especially important for when you
#   have multiple panels) and NOT to the x- or y-axis.
fig, ax = plt.subplots()

# Create a simple line plot using the 'seaborn.lineplot' function.
sns.lineplot(

    # Set the input data frame.
    data = df,

    # Set the name of the column containing the data we want displayed
    # on the x-axis.
    x = "Year",

    # Set the name of the column containing the data we want displayed
    # on the y-axis.
    y = "Life_Expectancy",

    # Set the name of the column containing the data we want to use to
    # group the data by color.
    hue = "Country",

    # Set the palette to use.
    palette = custom_palette,

    # Set the type of marker we want to use to display the single data
    # points.
    marker = "o",

    # Set on which 'axis' (panel) we want to draw the plot.
    ax = ax,
)

# Set the title of the plot.
ax.set_title("Life expectancy over time")

# Set the label of the y-axis.
ax.set_ylabel("Life expectancy (years)")

# The markers take the same color as the line.

# You can see that the name of the x-axis is the name of the column
# of the data frame that contains the data displayed on the axis.

# You can also see that a legend is automatically displayed.

# We call the 'matplotlib.pyplot.show' function to display our plot.
plt.show()
```

### Scatter plot with trend lines

A scatter plot with a trendline overlays a fitted curve (often a straight line or a smooth curve like LOWESS) on top of the raw points. This combines **exploratory** and **model-based** views: you can see the variability in the data and a simple model that summarizes the main pattern. Itâ€™s commonly used to illustrate **linear relationships**, **direction and strength of association**, and to highlight where the data deviate from the fitted trend.

References:

- The [`seaborn.regplot`](https://seaborn.pydata.org/generated/seaborn.regplot.html) function.
References:
- The [`matplotlib.pyplot.subplots`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html) function.
- The [`matplotlib.axes.Axes.set_title`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html) function.
- The [`matplotlib.axes.Axes.set_xlabel`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html) function.
- The [`matplotlib.axes.Axes.set_ylabel`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylabel.html) function.
- A list of [all markers available in matplotlib](https://matplotlib.org/stable/api/markers_api.html).
- The color palettes available in [Matplotlib](https://matplotlib.org/stable/users/explain/colors/colormaps.html) or [Seaborn](https://seaborn.pydata.org/generated/seaborn.color_palette.html#seaborn.color_palette).

```python
# Select only data for a specific year.
df = healthexp[healthexp["Year"] == 2020]

# Use the 'matplotlib.pyplot.subplots' function to create:
# - The 'figure' (fig), aka the full plot area.
# - The 'axis' (ax), which, a bit counterintuitively, refers to a
#   single panel within the figure (especially important for when you
#   have multiple panels) and NOT to the x- or y-axis.
fig, ax = plt.subplots()

# This time, use the 'seaborn.regplot' function to produce scatterplot with trendlines.
sns.regplot(

    # Set the input data frame.
    data = df,

    # Set the name of the column containing the data we want displayed
    # on the x-axis.
    x = "Life_Expectancy",

    # Set the name of the column containing the data we want displayed
    # on the y-axis.
    y = "Spending_USD",

    # Set a dictionary of arguments ("keywords") setting the properties
    # of the scatterplot's markers.
    scatter_kws = {"s" : 80},

    # Set a dictionary of arguments ("keywords") setting the properties
    # of the lines.
    line_kws = {"linewidth" : 2},

    # Set on which 'axis' (panel) we want to draw the plot.
    ax = ax
)

# Set the title of the plot.
ax.set_title("OLS trend line: spending vs life expectancy (2020)")

# Set the label of the x-axis.
ax.set_xlabel("Life expectancy (years)")

# Set the label of the y-axis.
ax.set_ylabel("Health spending per capita (USD)")

# Call the 'matplotlib.pyplot.show' function to display our plot.
plt.show()
```

### Scatterplot with LOWESS-style smoothing

Here, we smooth the time trend of life expectancy for our three example countries.

References:

## Bar charts

Bar charts represent categorical data with rectangular bars whose lengths are proportional to the values they encode. Theyâ€™re ideal for comparing **counts**, **totals**, or **averages** across discrete categories (e.g. spending by country in a given year). Vertical bars emphasize comparison across categories; horizontal bars are often better when labels are long or there are many categories. Grouped/stacked variants show composition or side-by-side comparisons.

```python
# Select only data for a specific year.
df_2020 = healthexp[healthexp["Year"] == 2020]

# Use the 'matplotlib.pyplot.subplots' function to create:
# - The 'figure' (fig), aka the full plot area.
# - The 'axis' (ax), which, a bit counterintuitively, refers to a
#   single panel within the figure (especially important for when you
#   have multiple panels) and NOT to the x- or y-axis.
fig, ax = plt.subplots()

# Create the bar plot using the 'seaborn.barplot' function.
sns.barplot(

    # Set the input data frame.
    data = df_2020,

    # Set the name of the column containing the data we want displayed
    # on the x-axis.
    x = "Country",

    # Set the name of the column containing the data we want displayed
    # on the y-axis.
    y = "Spending_USD",

    # Set the name of the column containing the data we want to use to
    # group the data by color.
    hue = "Country",

    # Set on which 'axis' (panel) we want to draw the plot.
    ax = ax
)

# For each bar in the bar plot
for p in ax.patches:

    # Get the current height of the bar.
    height = p.get_height()

    # Add a text element (the value the bar reaches in height)
    ax.text(

        # Obtain the position of the current text element on the
        # x-axis by taking the position of the current bar and adding
        # half of the bar's width (so that the text is centered on the
        # bar)
        x = p.get_x() + p.get_width() / 2,

        # Obtain the position of the current text element on the
        # y-axis by taking the height of a bar and adding a fraction
        # of it on top (so that there is a bit of space between the
        # bar and the text).
        y = height + (height * 0.05),

        # Set the text to be displayed (a rounded version of the
        # actual value).
        s = f"{height:,.0f}",

        # Set the horizontal alignment of the text so that it is
        # centered.
        ha = "center",

        # Set the vertical alignment of the text so that it is as
        # close as possible to the bar itself (apart from the padding).
        va = "bottom",

        # Set the size of the font.
        fontsize = 8,

        # Rotate the text 90 degrees.
        rotation = 90,
    )

# Set the title of the plot.
ax.set_title("Health spending per person (2020)")

# Set the label of the y-axis.
ax.set_ylabel("Health spending per capita (USD)")

# We need to extend the length of the y-axis a bit so that the text
# on top of the bar stays inside the plot's frame.

# Get the height of the tallest bar.
max_height = max(p.get_height() for p in ax.patches)

# Set a new limit for the y-axis.
ax.set_ylim(0, max_height * 1.5)

# Call the 'matplotlib.pyplot.show' function to display our plot.
plt.show()
```

Theyâ€™re especially useful for comparing distributions **across groups** (e.g. total bill by smoker vs non-smoker). Box plots emphasize differences in **central tendency**, **spread**, and **symmetry**, while hiding individual data points, which makes them compact but less detailed than scatter-based views.

References:

References:
- The [`matplotlib.pyplot.subplots`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html) function.

```python
# Use the 'matplotlib.pyplot.subplots' function to create:
# - The 'figure' (fig), aka the full plot area.
# - The 'axis' (ax), which, a bit counterintuitively, refers to a
#   single panel within the figure (especially important for when you
#   have multiple panels) and NOT to the x- or y-axis.
fig, ax = plt.subplots()

# Use the 'seaborn.boxplot' function to produce a box plot.
sns.boxplot(

    # Set the input data frame.
    data = tips,

    # Set the name of the column containing the data we want displayed
    # on the x-axis.
    x = "smoker",

    # Set the name of the column containing the data we want displayed
    # on the x-axis.
    y = "total_bill",

    # Set the name of the column containing the data we want to use to
    # group the data by color.
    hue = "sex",

    # Set on which 'axis' (panel) we want to draw the plot.
    ax = ax
)

# Set the title of the plot.
ax.set_title("Total bill by smoker and sex")

# Call the 'matplotlib.pyplot.show' function to display our plot.
plt.show()
```

```python
# Use the 'matplotlib.pyplot.subplots' function to create:
# - The 'figure' (fig), aka the full plot area.
# - The 'axis' (ax), which, a bit counterintuitively, refers to a
#   single panel within the figure (especially important for when you
#   have multiple panels) and NOT to the x- or y-axis.
fig, ax = plt.subplots()

# Use the 'seaborn.boxplot' function to produce a box plot.
sns.boxplot(

    # Set the input data frame.
    data = tips,

    # Set the name of the column containing the data we want displayed
    # on the x-axis.
    x = "smoker",

    # Set the name of the column containing the data we want displayed
    # on the x-axis.
    y = "total_bill",

    # Set the name of the column containing the data we want to use to
    # group the data by color.
    hue = "sex",

    # Set the palette to be used for coloring.
    palette = "husl",

    # Set on which 'axis' (panel) we want to draw the plot.
    ax = ax,

    # Set a dictionary of properties for the boxes.
    boxprops = {"alpha" : 0.5},

    # Set a dictionary of properties for the fliers.
    flierprops = {"alpha" : 0.5},

    # Set a dictionary of properties for the median lines.
    medianprops = {"alpha" : 0.5},

    # Set a dictionary of properties for the caps.
    capprops = {"alpha" : 0.5},

    # Set a list of properties for the whiskers.
    whiskerprops = {"alpha" : 0.5},
)

# Add a strip plot showing all the data points overlaid to the
# corresponding box plots.
sns.stripplot(

    # Set the input data frame.
    data = tips,

    # Set the name of the column containing the data we want displayed
    # on the x-axis.
    x = "smoker",

    # Set the name of the column containing the data we want displayed
    # on the x-axis.
    y = "total_bill",

    # Set the name of the column containing the data we want to use to
    # group the data by color.
    hue = "sex",

    # Set the palette to be used for coloring.
    palette = "husl",

    # Separate the "strips" of points by the same groups identified
    # by the column set in "hue".
    dodge = True,

    # Make the markers a bit transparent.
    alpha = 0.6,

    # Set the size of the markers.
    size = 3,

    # Set on which 'axis' (panel) we want to draw the plot.
    ax = ax
)

# Get the position and text of the different legend entries.
handles, labels = ax.get_legend_handles_labels()

# Convert it into a dictionary mapping each label's text to the label's
# position.
by_label = dict(zip(labels, handles))

# Plot the legend.
ax.legend(by_label.values(),
          by_label.keys(),
          title = "Sex")

# Set the title of the plot.
ax.set_title("Total bill by smoker and sex")

# Set the label of the x-axis.
ax.set_xlabel("Smoker")

# Set the label of the y-axis.
ax.set_ylabel("Total bill in USD")

# Call the 'matplotlib.pyplot.show' function to display our plot.
plt.show()
```

### Violin plot

Violin plots combine ideas from box plots and kernel density plots. Each â€œviolinâ€ shows a smoothed estimate of the distributionâ€™s shape, often with a central marker for the median and sometimes a mini box plot. Theyâ€™re helpful when you want to compare **full distribution shapes** across groups (e.g. bimodality or heavy tails), not just medians and quartiles. They work best with enough data to support a meaningful density estimate.

```python
# Use the 'matplotlib.pyplot.subplots' function to create:
# - The 'figure' (fig), aka the full plot area.
# - The 'axis' (ax), which, a bit counterintuitively, refers to a
#   single panel within the figure (especially important for when you
#   have multiple panels) and NOT to the x- or y-axis.
fig, ax = plt.subplots()

# Use the 'seaborn.violinplot' function to create the plot.
sns.violinplot(

    # Set the input data frame.
    data = tips,

    # Set the name of the column containing the data we want displayed
    # on the x-axis.
    x = "smoker",

    # Set the name of the column containing the data we want displayed
    # on the x-axis.
    y = "total_bill",

    # Set the name of the column containing the data we want to use to
    # group the data by color.
    hue = "sex",

    # Separate the "strips" of points by the same groups identified
    # by the column set in "hue".
    dodge = True,

    # Set what to display inside the violins.
    inner = "quartile",

    # Set on which 'axis' (panel) we want to draw the plot.
    ax = ax
)

# Set the title of the plot.
ax.set_title("Total bill by smoker and sex")

# Set the label of the x-axis.
ax.set_xlabel("Smoker")

# Set the label of the y-axis.
ax.set_ylabel("Total bill in USD")

# Set the title of the legend.
ax.legend(title = "Sex")

# Call the 'matplotlib.pyplot.show' function to display our plot.
plt.show()
```

This design lets you see both the **joint relationship** (how x and y move together) and the **individual distributions** of each variable. ## Heatmaps (correlation matrix)

Heatmaps encode values in a matrix or grid as colors, allowing you to see patterns in two-dimensional structured data at a glance. A common use in data analysis is a **correlation matrix heatmap**, where each cell shows the correlation between two variables and the color scale encodes the strength and sign. Heatmaps are great for spotting **blocks of high/low values**, **clusters**, and **overall patterns**, but individual values are usually read via annotations or the colorbar.

References:

- The [`seaborn.heatmap`](https://seaborn.pydata.org/generated/seaborn.heatmap.html) function.
- The [`matplotlib.pyplot.subplots`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html) function.
- The [`matplotlib.axes.Axes.set_title`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html) function.
- The color palettes available in [Matplotlib](https://matplotlib.org/stable/users/explain/colors/colormaps.html) or [Seaborn](https://seaborn.pydata.org/generated/seaborn.color_palette.html#seaborn.color_palette).

```python
# Select a list of countries of interest.
countries = ["USA", "Japan", "Germany"]

# Select only data relative to those countries.
df = healthexp[healthexp["Country"].isin(countries)]

# Compute the correlation between year/life expectancy/spending.
corr = df[["Year", "Life_Expectancy", "Spending_USD"]].corr()

# Just have a peek at how it looks like.
corr
```

```python
# Use the 'matplotlib.pyplot.subplots' function to create:
# - The 'figure' (fig), aka the full plot area.
# - The 'axis' (ax), which, a bit counterintuitively, refers to a
#   single panel within the figure (especially important for when you
#   have multiple panels) and NOT to the x- or y-axis.
fig, ax = plt.subplots()

# Use the 'seaborn.heatmap' function to draw the plot.
sns.heatmap(

    # Set the input data frame.
    data = corr,

    # Set whether to display the numerical value of each cell in the
    # cell itself.
    annot = True,

    # Set the format of the annotations (here, ".2f" means "a floating
    # point number with only 2 decimals after the point").
    fmt = ".2f",

    # Set the color map (palette) to use.
    cmap = "RdBu_r",

    # Set where the "center" of the color map is located.
    center = 0,

    # Set the minimum value represented by the color map.
    vmin = -1,

    # Set the maximum value represented by the color map.
    vmax = 1,

    # Set on which 'axis' (panel) we want to draw the plot.
    ax = ax

)

# Call the 'matplotlib.pyplot.show' function to display our plot.
plt.show()
```

## Using custom fonts for text elements

Matplotlib uses a default font, but you can fully customize fonts with
[`matplotlib.font_manager.FontProperties`](https://matplotlib.org/stable/api/font_manager_api.html#matplotlib.font_manager.FontProperties).

Key points:

- `family`, `weight`, `style`, `size` control the style.
- `fname` lets you load a **specific font file** (e.g. a `.ttf`).

You can use `matplotlib.font_manager.findSystemFonts()` to list available font files.

```python
# Import the required function.
from matplotlib.font_manager import findSystemFonts

# Get the font files available in your system.
font_files = findSystemFonts()

# Inspect some of the paths.
font_files[:5]
```

```python
# Import the 'FontProperties' class.
from matplotlib.font_manager import FontProperties

# Replace this with a path that exists on your system.
font_path = font_files[0]

# Set the font properties for the title.
title_font = FontProperties(fname=font_path,
                            size = 16,
                            weight = "bold")

# Set the font properties for the labels
label_font = FontProperties(fname=font_path,
                            size = 12)

# Create the figure and panel(s).
fig, ax = plt.subplots()

# Draw a scatter plot.
sns.scatterplot(data = tips,
                x = "total_bill",
                y = "tip",
                ax = ax)

# Set the title of the plot.
ax.set_title("Custom font for title",
             fontproperties = title_font)

# Set the label for the x-axis.
ax.set_xlabel("Total bill",
              fontproperties = label_font)

# Set the label for the y-axis.
ax.set_ylabel("Tip",
              fontproperties = label_font)

# Show the plot.
plt.show()
```

## The hierarchical structure of Matplotlib

Matplotlib is built around a small hierarchy of Python objects that together describe everything you see in a plot. At the top is the [`matplotlib.figure.Figure`](https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.html) class, which is the top-level container for all plot elements in a single â€œpageâ€ or window. A [`Figure`](https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.html) can contain one or more subplots, insets, color bars, and annotations, and you usually create it via helper functions like [`matplotlib.pyplot.figure`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html) or [`matplotlib.pyplot.subplots`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html).

Inside a [`Figure`](https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.html) you place one or more [`matplotlib.axes.Axes`](https://matplotlib.org/stable/api/axes_api.html#matplotlib.axes.Axes) objects. An [`Axes`](https://matplotlib.org/stable/api/axes_api.html#matplotlib.axes.Axes) represents one plot region (a single subplot), and it owns the x- and y-axis, the tick marks and labels, the title, the legend, and all the data that are actually drawn.

Under the hood, almost everything you see in a [`Figure`](https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.html) or [`Axes`](https://matplotlib.org/stable/api/axes_api.html#matplotlib.axes.Axes) is an [`matplotlib.artist.Artist`](https://matplotlib.org/stable/api/artist_api.html#matplotlib.artist.Artist). [`Artist`](https://matplotlib.org/stable/api/artist_api.html#matplotlib.artist.Artist) is an abstract base class for all drawable things: lines, markers, patches, text, images, legends, and even the [`Figure`](https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.html)  and [`Axes`](https://matplotlib.org/stable/api/axes_api.html#matplotlib.axes.Axes) themselves are subclasses of [`Artist`](https://matplotlib.org/stable/api/artist_api.html#matplotlib.artist.Artist). This means they all share a common interface for properties like `alpha`, `zorder`, etc.

Within an [`Axes`](https://matplotlib.org/stable/api/axes_api.html#matplotlib.axes.Axes), you typically interact with more specific Artist subclasses. Lines are [`matplotlib.lines.Line2D`](https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html) objects; bar charts and boxes are various [`matplotlib.patches.Patch`](https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Patch.html) subclasses; labels and titles are [`matplotlib.text.Text`](https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text) objects; legends are [`matplotlib.axes.Axes.legend`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.legend.html#matplotlib.axes.Axes.legend) objects, and collections of many similar shapes are [`Collection` subclasses](https://matplotlib.org/stable/api/collections_api.html). All of these inherit from [`matplotlib.artist.Artist`](https://matplotlib.org/stable/api/artist_api.html#matplotlib.artist.Artist), so they support both their own specific properties *and* the generic Artist properties, even if the function you called does not explicitly list every possible option.

```python
# Create a figure and the panel(s).
fig, ax = plt.subplots()

# Plot two lines
line1 = ax.plot([0, 1, 2], [1, 2, 0], label="line 1")
line2 = ax.plot([0, 1, 2], [0, 1, 0], label="line 2")

# Set the title of the plot.
ax.set_title("Figure / Axes / Artist hierarchy demo")

# Set the labels of the two axes.
ax.set_xlabel("x")
ax.set_ylabel("y")

# Set the legend.
ax.legend()

# Print useful info.
print("Figure:", fig)
print("Axes in figure:", fig.axes)
print("Lines in ax:", ax.lines)
```

### Setting properties on different levels

- **Figure-level**: `fig.suptitle`, size, facecolor  
- **Axes-level**: titles, labels, grids, axis limits, tick params  
- **Artist-level**: line color, linewidth, alpha, marker, etc.

```python
# Figure-level.

# Create a figure and the panel(s) (specify the size of the figure
# in inches and how many rows/columns should be in the figure's
# layout).
fig, axes = plt.subplots(nrows = 1,
                         ncols = 2,
                         figsize=(8, 3))

# For each available panel
for i, ax in enumerate(axes):

    # Plot a line.
    ax.plot([0, 1, 2], [0, i, 2*i], label=f"line {i}")

    # Plot the legend.
    ax.legend()

# Set the title for the whole figure.
fig.suptitle("Figure-wide title",
             fontsize = 16)

# Set the figure's background color.
fig.set_facecolor("0.95")

# Show the plot.
plt.show()
```

```python
# Axis-level.

# Create a figure and the panel(s).
fig, ax = plt.subplots()

# Draw a scatter plot.
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    ax=ax
)

# Set the panel's title.
ax.set_title("Axes-level settings")

# Set the limits for the x-axis.
ax.set_xlim(0, 60)

# Set the limits for the y-axis.
ax.set_ylim(0, 12)

# Enable a background grid on both axis.
ax.grid(True,
        which = "both",
        linestyle = "--",
        alpha = 0.5)

# Rotate the ticks' labels of the x-axis.
ax.tick_params(axis = "x",
               labelrotation = 45)

# Show the plot.
plt.show()
```

```python
# Artist-level.

# Create a figure and the panel(s).
fig, ax = plt.subplots()

# Plot two lines.
lines = ax.plot([0, 1, 2],
                [1, 2, 0],
                [0, 1, 2],
                [0, 1, 0])

# `lines` is a list of Line2D artists.
# Line2D inherits from Artist, so we can set generic Artist properties:
for line in lines:
    line.set_alpha(0.5)    # Artist property
    line.set_linewidth(3)

# Set the plot's title.
ax.set_title("Artist properties (including inherited ones)")

# Show the plot.
plt.show()
```
